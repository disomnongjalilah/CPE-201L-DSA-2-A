```python
def traverse_array(arr):
    """Function to traverse and print all elements in the array"""
    print("Traversing the array:")
    for i, element in enumerate(arr):
        print(f"Index {i}: {element}")

def reverse(arr):
    print("Reversing all elements")
    arr.reverse()
    print("Array after reversing:", arr)

def sort(arr):
    print("Sorting all elements")
    arr.sort()
    print("Array after sorting:", arr)

def insert(arr):
    try:
        value = int(input("Enter value to insert: "))
        position = int(input("Enter position to insert at: "))
        if position < 0 or position > len(arr):
            print("Invalid position.")
        else:
            arr.insert(position, value)
            print("Array after insertion:", arr)
    except ValueError:
        print("Invalid input.")

def delete(arr):
    try:
        value = int(input("Enter value to delete: "))
        if value in arr:
            arr.remove(value)
            print("Array after deletion:", arr)
        else:
            print("Value not found in array.")
    except ValueError:
        print("Invalid input.")

def modify(arr):
    try:
        index = int(input("Enter index to modify: "))
        if 0 <= index < len(arr):
            new_value = int(input("Enter new value: "))
            arr[index] = new_value
            print("Array after modification:", arr)
        else:
            print("Invalid index.")
    except ValueError:
        print("Invalid input.")

def search(arr):
    try:
        value = int(input("Enter value to search for: "))
        if value in arr:
            print(f"{value} found in the array.")
        else:
            print(f"{value} not found in the array.")
    except ValueError:
        print("Invalid input.")

def append(arr):
    try:
        value = int(input("Enter value to append: "))
        arr.append(value)
        print("Array after appending:", arr)
    except ValueError:
        print("Invalid input.")

def pop(arr):
    try:
        index = int(input("Enter index to pop: "))
        if 0 <= index < len(arr):
            popped_value = arr.pop(index)
            print(f"Popped value: {popped_value}")
            print("Array after popping:", arr)
        else:
            print("Invalid index.")
    except ValueError:
        print("Invalid input.")


def index(arr):
    try:
        value = int(input("Enter value to find index of: "))
        if value in arr:
            index = arr.index(value)
            print(f"Index of {value}: {index}")
        else:
            print(f"{value} not found in the array.")
    except ValueError:
        print("Invalid input.")

def count(arr):
    try:
        value = int(input("Enter value to count: "))
        count = arr.count(value)
        print(f"Count of {value}: {count}")
    except ValueError:
        print("Invalid input.")

def clear(arr):
    arr.clear()
    print("Array after clearing:", arr)


def main():
    # Initialize an array
    arr = [10, 20, 30, 40, 50]
    print("Original array:", arr)

    while True:
        print("\nMenu:")
        print("1. Traverse array")
        print("2. Reverse array")
        print("3. Sort array")
        print("4. Insert")
        print("5. Delete")
        print("6. Modify")
        print("7. Search")
        print("8. Append")
        print("9. Pop")
        print("10. Index")
        print("11. Count")
        print("12. Clear")
        print("13. Exit")

        choice = input("Enter your choice (1-13): ")

        if choice == '1':
            traverse_array(arr)
        elif choice == '2':
            reverse(arr)
        elif choice == '3':
            sort(arr)
        elif choice == '4':
            insert(arr)
        elif choice == '5':
            delete(arr)
        elif choice == '6':
            modify(arr)
        elif choice == '7':
            search(arr)
        elif choice == '8':
            append(arr)
        elif choice == '9':
            pop(arr)
        elif choice == '10':
            index(arr)
        elif choice == '11':
            count(arr)
        elif choice == '12':
            clear(arr)
        elif choice == '13':
            print("Exiting the program.")


if __name__ == "__main__":
    main()
```
